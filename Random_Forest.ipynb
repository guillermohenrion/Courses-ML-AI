# Random Forest: Introducción y Ejemplo Práctico

## 1. Introducción a Random Forest

Random Forest es un algoritmo de aprendizaje supervisado que se utiliza tanto para clasificación como para regresión. Se basa en la construcción de múltiples árboles de decisión durante el entrenamiento y la salida de la clase que es el modo de las clases (clasificación) o la media de las predicciones (regresión) de los árboles individuales.

### ¿Por qué Random Forest?

1. **Reducción de la varianza:** Combina las predicciones de múltiples árboles de decisión para mejorar la precisión y controlar el sobreajuste.
2. **Robustez:** Es menos sensible a los datos ruidosos y outliers.
3. **Importancia de las características:** Permite identificar las características más importantes en el conjunto de datos.

## 2. Conceptos Clave

### Árboles de Decisión

Un árbol de decisión es un modelo predictivo que divide el espacio de características en regiones rectangulares basándose en la condición de las características. Cada nodo interno representa una condición en un atributo, cada rama representa el resultado de la condición y cada hoja representa una etiqueta de clase o un valor de regresión.

### Ensamblado de Árboles

Random Forest construye varios árboles de decisión utilizando diferentes subconjuntos de datos y características, combinando sus resultados para mejorar la precisión general.

### Hiperparámetros Importantes

1. `n_estimators`: Número de árboles en el bosque.
2. `max_depth`: Profundidad máxima de los árboles.
3. `min_samples_split`: Número mínimo de muestras necesarias para dividir un nodo.
4. `min_samples_leaf`: Número mínimo de muestras que debe tener un nodo hoja.
5. `max_features`: Número máximo de características consideradas para dividir un nodo.

## 3. Ejemplo Práctico

Vamos a ilustrar el uso de Random Forest con un ejemplo práctico utilizando el conjunto de datos Iris.


```python
# Importar bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from mlxtend.plotting import plot_decision_regions

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data[:, :2]  # Usaremos solo las dos primeras características para facilitar la visualización
y = iris.target

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar un árbol de decisión
tree_clf = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_clf.fit(X_train, y_train)

# Crear y entrenar un Random Forest
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
rf_clf.fit(X_train, y_train)

# Evaluar los modelos
tree_accuracy = tree_clf.score(X_test, y_test)
rf_accuracy = rf_clf.score(X_test, y_test)
print(f'Exactitud del Árbol de Decisión: {tree_accuracy:.2f}')
print(f'Exactitud del Random Forest: {rf_accuracy:.2f}')
```

```python
# Definir el tamaño de la figura
plt.figure(figsize=(12, 5))

# Árbol de Decisión
plt.subplot(1, 2, 1)
plot_decision_regions(X, y, clf=tree_clf, legend=2)
plt.title('Árbol de Decisión')

# Random Forest
plt.subplot(1, 2, 2)
plot_decision_regions(X, y, clf=rf_clf, legend=2)
plt.title('Random Forest')

plt.show()
```
